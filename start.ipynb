# Tensor

Columns vs Depth

1. Columns (2D Context)
When dealing with a Rank-2 tensor (a matrix), the structure is defined by (Rows, Columns).

Axis: Usually Axis 1.
Visual: Think of a spreadsheet. The columns run vertically, but you count them across the horizontal axis.
Use Case: Features in a dataset, pixels in a single row of a grayscale image.


2. Depth (3D Context)
When dealing with a Rank-3 tensor (a cube or volume), the structure is often (Height, Width, Depth) or (Rows, Columns, Depth).

Axis: Usually Axis 2 (in "channels_last" convention).
Visual: Think of a stack of spreadsheets or matrices behind one another.
Use Case:
Color Channels: In an RGB image, the depth is 3 (Red, Green, Blue layers).
Filters: In Convolutional Neural Networks (CNNs), depth represents the number of feature maps.
# 3d tensor image like structure
3. Height vs. Depth
Height: A spatial dimension. It defines the physical size or resolution of the data vertically.
