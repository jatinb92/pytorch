```python
import torch
import torch.nn as nn
```

```python
class Model(nn.Module):
    
    def __init__(self, num_features, hidden_layer1_size):
        super().__init__()
        
        
        self.num_features = num_features
        self.hidden_layer1_size = hidden_layer1_size
        
        self.neural_network = nn.Sequential(
            nn.Linear(num_features, hidden_layer1_size),
            nn.ReLU(),
            nn.Linear(hidden_layer1_size, 1),
            nn.Sigmoid()
        )
      
        
        
        
    def forward(self, features):
        # print(f"features : {features}")
        # linear1_out = self.hidden_layer(features) # y = Wx + b
        
        # print(f"linear1_out : {linear1_out} ")
        # activation1_out = self.relu_activation(linear1_out)
        
        # linear2_out = self.linear2(activation1_out) # y = Wx + b
        activation2_out = self.neural_network(features)

        print(f"features : {features}")
        # linear1_out = self.hidden_layer(features) # y = Wx + b
        
        # print(f"linear1_out : {linear1_out} ")
        # activation1_out = self.relu_activation(linear1_out)
        
        # linear2_out = self.linear2(activation1_out) # y = Wx + b
        return activation2_out
```

```python
loss_function = nn.BCELoss() # binary cross entropy loss
```

```python
# fake dataset

no_of_features = 5
hidden_layer_size = 3

features = torch.rand(10, no_of_features)


# model
model = Model(features.shape[1], hidden_layer_size)


# internally NN Module calls forward method
y_pred = model(features) # same as model.forward(features)




print(y_pred)
```
