```python
import torch
```

```python
x = torch.tensor(3.0, requires_grad=True) # by default requires_grad is False
```

```python
y = x**2 

print(y) # tensor(9., grad_fn=<PowBackward0>) - gradient - powerbackward is the function to generate the input's derivative.
```

```python
# to find the gradient of y with respect to x, we use backward()

y.backward() 
```
